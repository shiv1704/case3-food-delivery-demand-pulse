```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("../data/raw/case3_food_delivery_orders.csv")
```

```python
df.head()
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.info()
```

```python
df.to_csv("../data/processed/cleaned_orders.csv", index=False)
```

```python
df['hour'] = df['timestamp'].dt.hour
```

```python
df['weekday'] = df['timestamp'].dt.day_name()
```

```python
df['is_weekend'] = df['weekday'].isin(['Saturday', 'Sunday'])
```

```python
def meal_period(hour):
    if 6 <= hour < 11:
        return "Breakfast"
    elif 11 <= hour < 16:
        return "Lunch"
    elif 16 <= hour < 20:
        return "Snacks"
    else:
        return "Dinner"

df['meal_period'] = df['hour'].apply(meal_period)
```

```python
df.to_csv("../data/processed/final_orders.csv", index=False)
```

```python
hourly_orders = df.groupby('hour').size().reset_index(name='orders')
```

```python
fig = px.line(
    hourly_orders,
    x='hour',
    y='orders',
    title='Orders by Hour'
)

fig.show()
```

```python
fig.write_image("../outputs/charts/Orders_by_Hour.png")
```

## Observations
- Order volumes remain consistently low during late-night and early-morning hours (12 AM–5 AM).
- Demand begins rising steadily from morning hours and accelerates sharply around lunch time.
- A strong lunch peak is visible between 12 PM–1 PM.
- Evening demand is even stronger than lunch demand, with the highest order volumes occurring around 7 PM–8 PM.
- Demand declines rapidly after 9 PM, indicating the end of major ordering windows.

## Business Interpretation
- Customer ordering behavior is highly concentrated around meal-time windows.
- Dinner demand contributes the largest operational load and likely requires maximum rider availability.
- Early morning hours may experience rider underutilization if staffing remains constant throughout the day.

## Potential Action
- Introduce peak-hour rider scheduling focused on lunch and dinner windows.
- Allocate higher rider density during evening periods to reduce delivery delays.
- Optimize overnight staffing levels to avoid unnecessary operational costs during low-demand periods.

```python
surge_analysis = df.groupby('surge_applied')['delivery_time_min'].mean().reset_index()
```

```python
fig = px.bar(
    surge_analysis,
    x='surge_applied',
    y='delivery_time_min',
    title='Average Delivery Time: Surge vs Non-Surge'
)

fig.show()
```

```python
fig.write_image("../outputs/charts/Average_Delivery_Time:_Surge_vs_Non-Surge.png")
```

## Observations
- Orders with surge pricing enabled show an average delivery time of approximately 43.2 minutes.
- Non-surge orders have a lower average delivery time of around 39.5 minutes.
- Surge-applied orders therefore experience roughly 3–4 minutes higher delivery times on average.

## Business Interpretation
- Surge pricing appears to be activated during periods of operational stress and high demand.
- Despite surge activation, delivery times remain elevated, indicating that pricing interventions alone may not fully solve rider supply shortages.
- The current surge mechanism may be reactive rather than predictive, activating only after delivery delays begin increasing.

## Potential Action
- Combine surge pricing with proactive rider positioning and peak-hour staffing strategies.
- Introduce demand forecasting–based surge activation to anticipate operational pressure before delays escalate.
- Monitor high-delay surge windows separately to identify city-specific operational bottlenecks.

```python
fig = px.histogram(
    df,
    x='delivery_time_min',
    nbins=30,
    title='Delivery Time Distribution'
)

fig.show()
```

```python
fig.write_image("../outputs/charts/Delivery_Time_Distribution.png")
```

## Observations
- The majority of deliveries are completed within the 30–50 minute range.
- Peak delivery concentration appears around 35–45 minutes, indicating this is the platform’s most common fulfillment window.
- Delivery volumes decline significantly after 60 minutes, but a visible long-tail distribution extends beyond 80–100 minutes.
- A small number of extreme delivery delays may negatively impact customer satisfaction and platform reliability metrics.

## Business Interpretation
- Overall operational performance appears relatively stable for most deliveries.
- However, long-tail delivery delays suggest occasional operational inefficiencies such as rider shortages, traffic congestion, or peak-hour overloads.
- Even a small percentage of highly delayed orders can disproportionately impact customer retention and satisfaction.

## Potential Action
- Investigate high-delay outlier orders separately to identify operational root causes.
- Introduce SLA monitoring for deliveries exceeding 60 minutes.
- Use peak-hour demand forecasting and rider balancing to reduce long-tail delays during high-demand periods.

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date
```

```python
df.columns
```

```python
daily_orders = df.groupby('date').size().reset_index(name='y')

daily_orders.columns = ['ds', 'y']
```

```python
from prophet import Prophet

model = Prophet()
model.fit(daily_orders)
```

```python
future = model.make_future_dataframe(periods=7)
forecast = model.predict(future)
```

```python
fig = model.plot(forecast)
```

```python
forecast[['ds', 'yhat']].tail(7).to_csv(
    "../outputs/forecast/7_day_forecast.csv",
    index=False
)
```

## Observations
- Daily order demand remains relatively stable throughout the forecasting period, fluctuating mostly between 540–580 orders per day.
- The forecast shows recurring cyclical demand patterns, indicating consistent customer ordering behavior across weeks.
- No strong upward or downward long-term demand trend is visible during the observed period.
- Forecast uncertainty remains moderate, with confidence intervals generally ranging between approximately 520–590 orders.

## Business Interpretation
- Short-term food delivery demand appears predictable and operationally stable.
- Stable demand patterns make short-horizon forecasting suitable for rider scheduling and operational planning.
- Since no major volatility or rapid growth trend is observed, operational efficiency improvements may generate higher impact than aggressive scaling efforts in the short term.

## Potential Action
- Use rolling 7-day demand forecasts to optimize rider staffing during peak operational windows.
- Combine forecasting outputs with city-level demand heatmaps for more granular operational planning.
- Introduce forecast-driven surge activation to proactively manage high-demand periods before delivery delays increase.
